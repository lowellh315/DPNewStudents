# In this demo, we'll introduce you to matplotlib and seaborn, packages in python that are used to plot and visualize data. 

We'll first start by importing pyplot from matplotlib. When using matplotlib, we usually use pyplot to generate data visualizations. In addition to matplotlib, we'll also use the numpy and math packages for this tutorial. If you have python installed through anaconda, you should already have the package.

```python
from matplotlib import pyplot as plt
import numpy as np
```

For our first example, let's try plotting a quadratic function y = x^2 + 6x + 9 for 20 values of x ranging from -13 to 7. To do this, we first create a list/array of x values using the linspace function in matlab (for those that are not familiar with linspace, you essentially input three values into the function: (start value, end value, number of values). So, for our list, we'll use np.linspace(0, 10, 5)

```python
x = np.linspace(-13,7,20)
```

Next, we we'll create a function to represent our quadratic function:

```python
def quadratic(x):
    y = x**2 + 6*x + 9
    return y
```

Once we have our x-values and function, we can start plotting our function. To plot using pyplot, we'll use the command plt.plot (or pyplot.plot if you don't import pyplot as plt). Through plt.plot, we'll pass two inputs: (x-values, y-values). We plot our function below:

```python
plt.plot(x, quadratic(x))
```

As you can see, we have successfully plotted our function. Although visualizing data is the primary goal of matplotlib, our plot is not yet presentable. To be presentable, we would have to add a title, legend, x- and y-axis labels. Furthermore, we might also want to change the size of our plot. We show you how to do so here:

```python
# To add a label/legend, we pass a third input to our plot function:
# Let's name our label "function"
plt.plot(x, quadratic(x), label = 'function')

# To show a legend/label, we simply use plt.legend().
plt.legend()

# To add a title to our plot, we simply use plt.title("Insert title here"). Let's name our plot "Function Plot".
plt.title('Function Plot')

# To add x- and y-labels, we simply use plt.xlabel("Insert label here") and plt.ylabel("Insert label here"). 
# Let's name our x-label "x-values" and our y-label "y-values"
plt.xlabel('x-values')
plt.ylabel('y-values')

# To change plot size, we can use plt.figsize(width, height) or plt.rcParams['figure.figsize'] = (width, height)
# Both work well, but I prefer to use plt.rcParams (you might have to run the code twice for plot size to change)
# so let's use that to create a 20 by 10 plot:
plt.rcParams['figure.figsize'] = (20,10)

# Finally, let's show our plot:
plt.show()
```

Additionally, if we wanted to plot another function on the same plot we can simply add another plt.plot(x values, y values) line. Let's visualize this below using our original function and a new function (y = x^2 + 5x +6):

```python
def second_quadratic(x):
    y = x**2 + 5*x + 6
    return y
```

```python
# To add a label/legend, we pass a third input to our plot function:
# Let's name our first function's label "first function" and add "second function" for our second function
plt.plot(x, quadratic(x), label = 'first function')
plt.plot(x, second_quadratic(x), label = 'second function')

# To show a legend/labem, we simply use plt.legend().
plt.legend()

# To add a title to our plot, we simply use plt.title("Insert title here"). Let's name our plot "Function Plot".
plt.title('Function Plot')

# To add x- and y-labels, we simply use plt.xlabel("Insert label here") and plt.ylabel("Insert label here"). 
# Let's name our x-label "x-values" and our y-label "y-values"
plt.xlabel('x-values')
plt.ylabel('y-values')

# To change plot size, we can use plt.figsize(width, height) or plt.rcParams['figure.figsize'] = (width, height)
# Both work well, but I prefer to use plt.rcParams (you might have to run it twice for plot size to change)
# so let's use that to create a 20 by 10 plot:
plt.rcParams['figure.figsize'] = (20,10)

# Finally, let's show our plot:
plt.show()
```

Another interesting topic we'll look at is subplots: in matplotlib, we can plot and visualize multiple plots. This can be helpful when it's not particularly useful to visualize multiple data in one plot or if the x- and y-ranges do not correspond to each other well. Let's create a plot with four subplots. We already a quadratic function, so let's also create a line graph, cubic function, and a quartic function:

```python
def linear(x):
    y = 4*x + 7
    return y

def cubic(x):
    y = x**3 + 9*x**2 + 4*x
    return y

def quartic(x):
    y = 7*x**4 + 3*x - 4
    return y
```

```python
# To create subplots, we use the plt.subplots command. In this function, we pass the number of rows and columns
# of plots we want as (rows, columns)
fig, ax = plt.subplots(2,2)

# We can call and plot each specific subplot using the command ax[row, column]
# Thus, to plot our first subplot, we'll plot our function in the zeroth row and zeroth column (recall that python
# indexing begins at 0 and not 1)
# This wasn't noted earlier, but to change the color of the plot, we can pass another input to indicate our desired color
ax[0,0].plot(x, linear(x), color = 'blue')
ax[0,0].set_title('Linear Function')

# We repeat the following process for all four functions. 
ax[0,1].plot(x, quadratic(x), color = 'orange')
ax[0,1].set_title('Quadratic Function')
#ax[0,1].set(xlabel = 'x-values', ylabel = 'y-values')

ax[1,0].plot(x, cubic(x), color = 'green')
ax[1,0].set_title('Cubic Function')

ax[1,1].plot(x, quartic(x), color = 'purple')
ax[1,1].set_title('Quartic Function')

# Finally, to add in labels, we can use ax.set(xlabel = 'insert label', ylabel = 'insert label') 
# Since we want to use the same labels for all plots, we can also use the following for loop:
# This for loop goes through each subplot and sets it's labels to our choosing:
for i in ax.flat:
    i.set(xlabel = 'x-values', ylabel = 'y-values')
```

Okay!! So we now how to plot (multiple or single) functions/data using matplotlib as well as how to customize our plots. Let's now move on to an example involving pandas and how we can plot data from dataframes. In our first example, let's plot each student and their GPA. 

Our first order of business will be to import pandas and read the dataframe. To read the dataframe, make sure you know it's directory in your computer and make sure that this notebook is in a higher directory.

```python
import pandas as pd

gpa_data = pd.read_csv('dataset/dataset/education/grades.csv')
```

Let's first look at the dataframe, specifically the column names and data types. As you have seen from previous tutorials, to briefly view the dataframe, we use the command dataframe.head(n) where dataframe represents the name of your dataframe and n represents the rows of data you want to see.

```python
gpa_data.head(5)
```

```python
print(gpa_data.columns)
```

```python
print(gpa_data.dtypes)
```

Upon viewing the dataframe, we can see that the column 'uid' is object column and that the data we want to plot, the gpa all column, consists of floats and numerical values. This will be a problem since matplotlib cannot use strings as x-values. However, can resolve this issue. The first order of business will be to extract the ' gpa all' column and plot it. Since we can't use the 'uid' column as x-values, we'll use a placeholder. I typically use range(0, len(y-values)) so that the x-values range from 0 to the however large the set of data we're plotting is. Let's do this now:

```python
plt.plot(range(0, len(gpa_data[' gpa all'])), gpa_data[' gpa all'])
plt.show()
```

Once again, although we have plotted our data, the visualization is incomplete. Furthermore, we can't tell who's gpa is what and it's kinda hard to pin point the gpa with a line graph - this calls for a scatter plot and for a grid-style plot! Let's add these improvements now.

```python
# As usual, let's start by plotting the data. However, instead of a regular line plot, let's use a scatter plot.
# To plot a scatter plot, the only change to our command will be using plt.scatter instead of plt.plot.
# With the plt.scatter function, we can also indicate the size of our markers. Let's use s=100 for better 
# visualization.
plt.scatter(range(0, len(gpa_data[' gpa all'])), gpa_data[' gpa all'], s=100)

# Now let's change the x-ticks to student id. This will make it much easier to identify each student in the plot.
# To set the x-ticks, we use plt.xticks(x-values, insert ticks list here) and pass in a set of ticks. 
# x-values in the command will the original (placeholder) x-values. For us, it's range(0, len(gpa_data[' gpa all']))
# The ticks we want to use are the uids in the dataframe. Let's pass them in!
plt.xticks(range(0, len(gpa_data[' gpa all'])), list(gpa_data['uid']))

# Let's now add in x- and y-labels and a title.
plt.title('Student GPA')
plt.xlabel('Student ID')
plt.ylabel('GPA')

# Finally, we'll add in grids to make our plot easier to read:
# To add in grids, we use plt.grid(color = 'desired color', linestyle = 'desired linestyle')
plt.grid()

plt.show()
```

As you now observe, the figure looks much more polished and is easier to read and understand. This particular process was a bit different from before when we were working with x and y data that were purely numerical as we now worked with dataframes that often have a mix of strings and integers. This process is therefore especially useful when trying to plot data from dataframes. Although both graphs are equivalent, the second method of plotting becomes especially useful if you want to plot data from two different dataframes (i.e. if the x-values come from one dataframe and the y-values come from another dataframe). Let's look at an example: we'll stick with plotting student GPA, but let's vary the size of the markers by how the students scored on the PHQ-9 Depression Survey (in this example, a higher score = more severe symptoms of depression = larger marker size).

Let's first start by reading the PHQ-9 survey dataframe and merging this with the gpa dataframe. When working with two dataframes, it's always a good idea to merge them: if you, say, want to plot scores on the PHQ-9 survey on the x-axis and GPA on the y-axis, you might get an error since there might have been some students who reported their GPA but did not report/take the PHQ-9 (or vice-versa).

```python
phq9 = pd.read_csv('dataset/dataset/survey/PHQ-9.csv')
```

```python
phq9.head()
```

As you can see, the PHQ-9 dataframe has two types of surveys: pre and post. For our purposes, we'll be using the pre-survey. Let's only look at that data. Also, as you'll realize, although the PHQ-9 survey is supposed to return a score, we have object values that we cannot add up. To fix this, we can replace the object values with their corresponding scores (ranging from 0 for "Not at all" to 3 for "Nearly every day").

```python
phq9 = phq9[phq9['type'] == 'pre']
```

```python
def corresponding_score(data):
    columns = data.columns
    for i in columns:
        data[i] = data[i].replace('Not at all', 0)
        data[i] = data[i].replace('Several days', 1)
        data[i] = data[i].replace('More than half the days', 2)
        data[i] = data[i].replace('Nearly every day', 3)
        
    return data
```

```python
phq9_edited = corresponding_score(phq9)
```

Let's now have a look at our new dataframe:

```python
phq9_edited.head()
```

As you can see, there are still three columns that have object values, but these columns are not relevant for the PHQ-9 depression score. So, we can go ahead and find the scores for each student by summing each row (note: although each row has an object value, when summing each row, pandas ignores the object values for you):

```python
phq9_edited['phq9 score'] = phq9_edited.sum(axis=1)
```

Let's now have a look at our dataset:

```python
phq9_edited.head()
```

This looks good! Now, all we have to do is merge this dataframe with the gpa dataframe. Since we would only like to merge on two columns (uid and phq9 score), we can specify this:

```python
gpa_and_phq9 = pd.merge(gpa_data, phq9_edited[['uid', 'phq9 score']], how='inner', on='uid')
```

```python
gpa_and_phq9.head()
```

Okay!! So we've successfully merged the two dataframes. Although we have merged them, there are far too many unique PHQ-9 scores. However, we can fix this by assigning levels for the scores: since the PHQ-9 scores range from 0-27, we can make 4 'levels' of PHQ-9 scores: 0-7, 8-14, 15-21, and 22-27.

```python
def assign_phq9_labels(data):
    
    phq9_level = []
    for i in data['phq9 score']:
        
        if 0 <= i <= 7:
            phq9_level.append(1)
        if 8 <= i <= 14:
            phq9_level.append(2)
        if 15 <= i <= 21:
            phq9_level.append(3)
        if 22 <= i <= 27:
            phq9_level.append(4)
    
    data['phq9 level'] = phq9_level
    
    return data
```

```python
gpa_and_phq9_final = assign_phq9_labels(gpa_and_phq9)
```

```python
gpa_and_phq9_final
```

This dataframe looks good! Let's move to plotting this. The code is the exact same as before. The only difference? Instead of passing an integer for the size parameter, we pass the dataframe values:

```python
sns.set_style('whitegrid')
                                                               # This the only difference in the code.
sns.scatterplot(x = gpa_data['uid'], y = gpa_data[' gpa all'], s = gpa_and_phq9_final['phq9 level']*40)
                                                               # The only reason why we multiply by 40 is because 
                                                               # the size otherwise would be 1-4, which is too small to see
plt.title('Student GPA')
```

So, now when looking at the GPA of each student, we can get an idea of how depressed they were at the start of the semester. We can also have the same effect by instead of changing the size of the markers, we change the color. To do this, we change the hue of the plot. The code is still very much the same:

```python
sns.set_style('whitegrid')
                                                               
sns.scatterplot(x = gpa_data['uid'], y = gpa_data[' gpa all'], s = 100, hue = gpa_and_phq9_final['phq9 level'])
                                                              
plt.title('Student GPA')
```

Although there no clear correlations between GPA and PHQ-9 score, it does serve as a good example to show you how you can customize plots using seaborn to showcase data from your dataset more powerfully.
